Matplotlib is a Python library used for 2D and 3D graphics. Graphical representations are very useful in data work because they allow us to better understand the data and to more finely track some algorithm behaviors.

To be able to use the matplotlib library, we first need to import it. We will do this with the following command.

```python
import matplotlib.pyplot as plt
```

We will also import the `numpy` library as we will need it for creating sets of points for display.

```python
import numpy as np
```

## Plotting Function Graphs

In this example, we will plot the graph of the function $y = \sin(x)$ in the interval $[0, 10]$.

First, we will form a dense grid of points along the x-axis in the interval [0, 10] that are evenly spaced (such a grid is called an equidistant grid) so that our graph looks fine. We will do this by calling the `linspace` function from the numpy library. This function expects the interval endpoints 0 and 10 as arguments and the number of division points - in our case, this can be the number 100.

```python
x = np.linspace(0, 10, 100)
```

Then we will calculate the value of the sine function for each of these points by calling the `sin(x)` function. We will save its values in the variable `y` - and it will now be an array of 100 points because the function is applied to each element of the array `x`.


```python
y = np.sin(x)
```

Now we will plot the graph and explain what each part means.

```python
# setting the title of the graph and axis labels
plt.title('Graph of the function y=sin(x)')
plt.xlabel('x')
plt.ylabel('y')

# plotting the graph
plt.plot(x, y)

# displaying the graph
plt.show()
```

We set the graph title by calling the `title` function, and the axis labels by calling the `xlabel` and `ylabel` functions. All these functions are defined at the `plt` drawing panel level. The graph itself is drawn by calling the `plot` function and specifying the `x` and `y` coordinate values for drawing (that's why we named the initial set of points and the sine function values with these names). The graph is displayed by calling the `show` function.


## Drawing a Scatter Plot

```python
np.random.seed(7)
```

Scatter plots are often used to gain insight into the spatial distribution of data. In the following example, we will create ten pairs of points with integer coordinate values from the interval [0, 20] and display them as a scatter plot.


First, we will generate pairs of points by separately generating values for x and y. The `randomint` function from the numpy library will help us with this, whose arguments `low`, `high`, and `size` allow control of the lower and upper bounds of the interval, as well as control of the number of points.

```python
# arrays of length 10 with arbitrary elements from the interval [0, 20]
x = np.random.randint(low=0, high=20, size=10)
y = np.random.randint(low=0, high=20, size=10)
```

The scatter plot itself is created by calling the `scatter` function of the drawing panel `plt`. When calling this function, the coordinate values of the points are specified, in our case `x` and `y`. Additionally, the color of the points can be set with the `color` argument, and the symbol for display can be changed with the `marker` argument. Instead of the default black circles, green downward-pointing triangles are used on the graph. The graph is displayed, as in the previous example, by calling the `show` function.

```python
# generating scatter plot
plt.scatter(x, y, color='green', marker='v')

# displaying the graph
plt.show()
```

We can notice that in this example, at the level of the numpy library and its `random` package, we set the random number generator (the so-called `seed` property) to the value 7. This will allow us to get the same point distribution every time we run this code. This property is important to us because of the possibility of rerunning experiments.
